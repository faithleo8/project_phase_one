```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
%matplotlib inline
```

```python
budgets_dataframe = pd.read_csv("tn.movie_budgets.csv")
budgets_dataframe.head()
```

```python
columns_strip=["production_budget","domestic_gross","worldwide_gross"]
```

```python
budgets_dataframe[columns_strip] = budgets_dataframe[columns_strip].astype(str) 

budgets_dataframe[columns_strip]=budgets_dataframe[columns_strip].replace({'\$':'',',':''},regex=True)
budgets_dataframe[columns_strip] = budgets_dataframe[columns_strip].astype(float) 
budgets_dataframe
```

```python
budgets_dataframe.info()
```

```python
#drop columns with domestic gross equal to zero
new_budgets_dataframe=budgets_dataframe[budgets_dataframe["domestic_gross"]> 0]
new_budgets_dataframe
```

```python
new_budgets_dataframe.info()
```

```python
new_budgets_dataframe['Profit'] = new_budgets_dataframe['worldwide_gross'] - new_budgets_dataframe['production_budget']
new_budgets_dataframe.head()

new_budgets_dataframe['Profit_Margin'] = (new_budgets_dataframe['worldwide_gross'] - 
                                    new_budgets_dataframe['production_budget'])/new_budgets_dataframe['worldwide_gross']
new_budgets_dataframe.head()
```

```python
high_profit_movies_df = new_budgets_dataframe.loc[new_budgets_dataframe['Profit'] > 0]
high_profit_movies_df = new_budgets_dataframe.sort_values(by=['Profit'], ascending=False)
high_profit_movies_df.reset_index(inplace=True)
high_profit_movies_df=high_profit_movies_df[high_profit_movies_df["Profit_Margin"]> 0]
high_profit_movies_df.head(10)
```

```python
movie=high_profit_movies_df["movie"]
Profit=high_profit_movies_df["Profit"]
fig, ax = plt.subplots(figsize =(16, 9))
# Horizontal Bar Plot
ax.barh(movie[0:20], Profit[0:20])
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('movies vs profit ',
             loc ='left', )
 
# Add Text watermark
fig.text(0.9, 0.15, 'Jeeteshgavande30', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)
 
# Show Plot
plt.show()
plt.savefig('plot.png', dpi=300, bbox_inches='tight')
```

```python
'''
This graph shows the 20 most succesfful movies in relation to the budget.
This is a clear indication that the more the budget allocated for the movie, the more the profit.
'''
```

```python
#plotting a scatter plot to look at the relationship between the budget and the profit for movies
sns.lmplot(x='production_budget', y='Profit', data=high_profit_movies_df, height=7, aspect=2)
plt.xlabel('Budget in (Millions of Dollars)', fontsize=12)
plt.ticklabel_format(axis='x', style='sci', scilimits=(6,6))
plt.ylabel('Profit in (Billions of Dollars)', fontsize=12)
plt.title('Production Budget vs  Profit for Profitable Movies', fontsize=14);
```

```python
movie_genre_dataframe = pd.read_csv("title.basics.csv")
movie_genre_dataframe.head()
#movie_genre_dataframe.info()
```

```python
movie_genre_dataframe.info()
```

```python
movie_genre_dataframe.drop(labels = ['original_title','runtime_minutes','tconst'], axis = 1 ,inplace = True)
movie_genre_dataframe
```

```python
movie_genre_dataframe.info()
```

```python
movie_genre_dataframe.dropna(subset = ["genres"], inplace=True)
```

```python
#merging the budget and genres table to look at the genres that produce the highest profit
movie_buget_and_genre_merge=pd.merge(movie_genre_dataframe,high_profit_movies_df,left_on='primary_title',right_on='movie')
movie_buget_and_genre_merge.tail()
```

```python
movie_buget_and_genre_merge = movie_buget_and_genre_merge.loc[movie_buget_and_genre_merge['Profit'] > 0]
movie_buget_and_genre_merge
```

```python
movie_buget_and_genre_merge.info()
```

```python
movie_buget_and_genre_merge.drop(labels =["primary_title",'id'],axis = 1, inplace =True)
#movie_buget_and_genre_merge2 = movie_buget_and_genre_merge.drop_duplicates(subset = "movie",keep = False,inplace = True)
movie_buget_and_genre_merge
```

```python
movie_buget_and_genre_merge.info()
```

```python
#movie_buget_and_genre_merge['genres'] = movie_buget_and_genre_merge['genres'].astype('str') 
movie_buget_and_genre_merge.duplicated()
```

```python
movie_buget_and_genre_merge['genres'] = movie_buget_and_genre_merge['genres'].str.split(',',n = 1, expand = True)
movie_buget_and_genre_merge.head()
```

```python
movie_buget_and_genre_merge = movie_buget_and_genre_merge.drop_duplicates()
movie_buget_and_genre_merge.head(30)
```

```python
movie_buget_and_genre_merge = movie_buget_and_genre_merge.drop_duplicates(subset=['index'])
movie_buget_and_genre_merge.head(30)
```

```python
movie_buget_and_genre_merge = movie_buget_and_genre_merge.dropna()
movie_buget_and_genre_merge.head(30)
```

```python
movie_buget_and_genre_merge1 = movie_buget_and_genre_merge.groupby('genres' , as_index=False)["movie"].count().sort_values(by='movie', ascending=False)
#.sort_values(by='movie', ascending=False)
#['movie'].count().sort_values(by='movie', ascending=False)
movie_buget_and_genre_merge.tail(30)
```

```python
#Plot the above findings.
plt.figure(figsize=(14,7))
ax1 = sns.barplot(x=movie_buget_and_genre_merge1['movie'], y=movie_buget_and_genre_merge1['genres'], palette='magma')
plt.xlabel('Movie Count', fontsize=12)
plt.ylabel('Genres', fontsize=12)
plt.title('Movie Count By Genre', fontsize=14);
```

```python
"""We can see that drama, comedy, and action dominate the quantity of movie genres but this doesnt necessarily mean these are the most profitable genres.
In order to determine this i will group each genre and take a look at the average net profit for each.
"""
```

```python
#Once again group the movies by genre, showing the average net profit and profit margin for each.
movie_buget_and_genre_merge3 = movie_buget_and_genre_merge.groupby('genres', as_index=False)['Profit'].median().sort_values(by='Profit',ascending = False)
movie_buget_and_genre_merge3.head()
```

```python
#Plot the above findings.
plt.figure(figsize=(14,7))
ax2 = sns.barplot(x=movie_buget_and_genre_merge3['Profit'], y=movie_buget_and_genre_merge3['genres'], palette='viridis')
plt.xlabel('Net Profit (Hundreds of Millions)', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.title('Net Profit By Genre', fontsize=14);
```
